```python
import numpy as np
import math
```

```python
from sklearn.cluster import KMeans
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_clean = pd.read_csv('Z_dataset_clean.csv')
df_clean
```

```python
matrix = df_clean.pivot_table(values=['trackname'],index=['user_id'],columns=['artistname'], aggfunc='count').fillna(0)
```

```python
matrix = (matrix/matrix).fillna(0).astype("int8")
```

```python
matrix
```

Метод "Локтя" для поиска оптимального количества кластеров:

```python
ssd = []

for k in range(2, 11):
    model = KMeans(n_clusters = k, max_iter=700)
    model.fit(matrix)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2, 11), ssd, 'o--')
plt.xlabel('Кол-во кластеров')
plt.ylabel('SSD') #Сумма квадратов расстояний от точек до центов кластеров
```

В виду проведенного исследования - определение оптимального количества кластеров (k), будет принято колличество равное 7. Был составлен цикл для обучения модели на разных k (от 2 до 10 включительно). График имеет убывающий характер (по большей части). Для дальнешего обучения будет принято значение k = 7. 
Как видно на графике, участок от 7 до 10, в точке 7 значение ошибки не значительно больше, чем в точке 8. Чтобы не сильно утяжелять модель, примем значение 7, т.к. процент ошибки более низкий, чем на участке 3-4.

```python
# Начинаем итоговое разбиение на кластеры

KM_model = KMeans(n_clusters=7, random_state=42, max_iter=700) # специально увеличила кол-во итераций
cluster_labels = KM_model.fit_predict(matrix)
cluster_labels
```

```python
# подсчет количества пользователей в каждом кластере
dct1 = {}
for i in cluster_labels:
    if i not in dct1:
        dct1[i] = 1
    else:
        dct1[i] += 1
dct1
```

Как видим, в некоторых кластерах оказался только 1 пользователь. В случае, поиска рекомендаций по трекам или авторам из такого кластера, составить рекомендацию из 10 треков становится маловозможным (если у пользователя не обширный плейлист). Если у пользователя будет менее 11 записей, то порекомендовать 10 треков становится невозможным. Необходимо определить, какой кластер находится ближе всего к указанному и при необходимости брать наиболее популярные треки оттуда.

p.s. В данной работе оставлены только пользователи с кол-ом треков не мнее 101. В таком случае, поиск ближайшего кластера не потребуется.

Ниже алгоритм поиска ближайшего кластера:

```python
# Координаты центроидов кластеров
centroid = KM_model.cluster_centers_
centroid
```

```python
# Разбиваем массив с центроидами построчно (каждый центр отдельно). Далее делаем список списков: [7 координат центров[координаты по осям каждого центра]].
centroid_arr = np.split(centroid, 7)
lst_coordinates = []
for elem in range(7):
    lst_coordinates.append([float(i) for i in list(centroid_arr[elem][0])])
# lst_coordinates
```

```python
mini_cluster = 4 #Сюда автоматически поступает значение из заданного условия (при необходимости)
lst_distance = []
new_cluster = 0
for i in range(7):
    if i != mini_cluster: # берутся только те координаты центров, которые не совпадают с заданным
        for j in range(len(lst_coordinates[mini_cluster])):
            sum_center = 0
            sum_center += math.sqrt(pow(lst_coordinates[i][j]-lst_coordinates[mini_cluster][j], 2)) # вычисляем сумму квадратов расстояний разницы координат заданного центра и остальных
        lst_distance.append(sum_center) # все суммы добавляем последовательно в список
```

```python
# ПИШЕТСЯ ОТДЕЛЬНЫЙ def...
# необходимо определить номер кластера, т.к. заданный кластер исключен из последовательности, произошел сдвиг на 1 позицию всех кластеров после.
if lst_distance.index(min(lst_distance)) >= mini_cluster: # выбираем минимальное значение расстояний из списка lst_distance и смотрим, номер индекса этого кластера в списке. Если он больше или равен заданному, то необходим номер кластера увеличить на 1. 
    new_cluster = lst_distance.index(min(lst_distance)) + 1
else:
    new_cluster = lst_distance.index(min(lst_distance)) # если нет, то оставляем текущий номер
new_cluster
```

ЗАКОНЧИЛСЯ АЛГОРИТМ ПОИСКА БЛИЖАЙШЕГО КЛАСТЕРА

```python
# добавление в датафрейм меток кластеров, далее выделяем нужные столбцы датафрейма из матрицы признаков в отдельные списки
# на основе списков составляем датафрейм для дальнейшей работы с ним
user_clusters = matrix.reset_index()[['user_id']]
user_clusters['cluster'] = cluster_labels
user = list(user_clusters['user_id'])
cluster = list(user_clusters['cluster'])
df_user_cluster = pd.DataFrame({'user_id': user, 'cluster': cluster})
# df_user_cluster
```

```python
# слияние исходного датафрейма с наименованиями кластеров 
# !!!!! ЗАМЕНИТЬ ПЕРЕМЕННУЮ df_user_cluster НА КОД pd.DataFrame({'user_id': user, 'cluster': cluster})
df_result = df_clean.merge(df_user_cluster, how= 'left', on= 'user_id')[['user_id', 'artistname', 'trackname', 'cluster']]
df_result
```

```python
# подсчет кол-ва треков исполнителя в каждом кластере
group = df_result.groupby(['cluster', 'artistname', 'trackname']).count().reset_index().rename(columns={'user_id': 'count_tracks'})
group
```

```python
# Подсчет количества раз добавления артиста в каждом кластере
group.groupby(['cluster', 'artistname'])[['count_tracks']].sum().reset_index().rename(columns= {'count_tracks': 'count_artist'})
```

```python
# слияние таблицы с подсчетом кол-ва треков и таблицы с подсчетом кол-ва раз добавленных артистов
df_recom = group.merge(group.groupby(['cluster', 'artistname'])[['count_tracks']].sum().reset_index().rename(columns= {'count_tracks': 'count_artist'}), 
            how='left', on=['cluster', 'artistname'])
df_recom
```
